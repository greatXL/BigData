```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
def fun1():
    x = np.arange(10)
    y = x
    #获取画板
    plt.figure()
    #x表示所有的x轴点
    #y表示所有的y轴点
    #"-o"表示线类型
    plt.plot(x,y,"-o",color="r")
fun1()
#?plt.plot
```

```python
def fun2():
    x = np.arange(10)
    y = x ** 2
    plt.figure()
    plt.plot(x, y, "--", color="b")
fun2()
```

```python
def fun3():
    #从1到10取100个数，这100个数为等差数列
    x = np.linspace(1,10,100)
    y = np.cos(x)
    z = np.sin(x)
    plt.figure()
    plt.plot(x, y,"-2")
    plt.plot(x, z, "-2")
fun3()
```

```python
def fun4():
    x = np.arange(10)
    y = x ** 2
    #设置画板宽8英寸，高4英寸
    plt.figure(figsize=(8,4))
    plt.plot(x, y,"-x", color="r")
    #x、y轴分别加标注，默认不支持中文
    plt.xlabel("xlabel")
    plt.ylabel("ylabel")
    
    #修改刻度
    plt.xticks(np.arange(0,10))
    #plt.xticks(np.arange(0,10),["a","b","c","d","e","f","g","h","i","j"])
    
    ax = plt.gca()#gca=> get current axis
    for i, j in zip(x,y):
        #ax.text()用来在坐标系中添加文字内容
        #text(x轴,y轴,文字内容....)
        ax.text(i+0.2,j,j,bbox=dict(facecolor='red', alpha=0.5))
    #背景添加表格
    plt.grid(True)
fun4()
#a = plt.gca()
#?a.text()
```

```python
def fun5():
    x = np.arange(50)
    #列表中存放6条线的不同颜色
    colors = ["r","y","g","b","c","m"]
    #字典中存放6条线的不同样式
    kind = {"circle":"o","start":"*","traggle_down":"v","dash_dot":"-","dash_line":"--","x-mark":"x"}
    plt.figure(figsize=(12,8))
    for index,key in zip(np.arange(6),kind.keys()):
        #符合正态分布的随机数
        y = np.random.randn(50).cumsum()
        plt.plot(x,y,kind[key],color=colors[index],label=key)\
    #legend()添加图例，图例中的内容名称跟plot函数中的label属性对应
    #legend()方法中有一个loc参数，可以用来设置图例的位置，默认是best
    plt.legend()
fun5()
```

```python
def fun6():
    plt.figure(figsize=(8,6))
    x = np.linspace(-np.pi,np.pi,256, endpoint=True)
    c = np.cos(x)
    s = np.sin(x)
    
    plt.plot(x, c, linewidth=2.5, linestyle="-", label="cos")
    plt.plot(x, s, linewidth=2.5, linestyle="--", label="sin")
    
    #指定x轴y轴的取值范围
    plt.xlim([-4,4])
    plt.ylim([-1,1])
    # r"$内容$"  斜体加转义
    plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r"$-\pi$",r"$-\pi/2$",r"$0$",r"$\pi/2$",r"$\pi$"])
    plt.yticks([-1,0,1])
    
    #去除上边框和右边框、
    ax = plt.gca()
    ax.spines["right"].set_color("none")
    ax.spines["top"].set_color("none")
    
    #将底边框，移到y轴0点位置
    ax.spines["bottom"].set_position(("data",0))
    #将左边框，移到x轴0点位置
    ax.spines["left"].set_position(("data",0))
    
    t = 2 * np.pi / 3
    y = np.sin(t)
    
    #画点使用scatter,50表示点的半径
    plt.scatter(t,y,50, color="r")
    plt.plot((t,t),(0,y),"--",color="r")
    
    #添加标注
    plt.annotate("y=sin(x)",xy=(t,y),textcoords="offset points",xytext=(30,30),arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.5"))
    
fun6()
```
