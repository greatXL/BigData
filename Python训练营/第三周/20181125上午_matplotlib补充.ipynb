```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
#直方图 特殊的柱状图
x = np.random.randint(0,10,10)
#orientation表示方向，默认是vertical
plt.hist(x,orientation="horizontal")
```

```python
#散点图
x = np.random.randn(1000)
y = np.random.randn(1000)
colors = np.random.random(3000).reshape((1000,3))
#s表示点的大小
plt.scatter(x,y,color=colors,s=10)
```

```python
def fun():
    plt.figure(figsize=(12,8))
    a1 = plt.subplot(2,3,1)
    x1 = np.arange(5)
    y1 = x1 + 5
    a1.plot(x1,y1)
    a1.grid(color="r",linestyle="--")
    
    a2 = plt.subplot(2,3,2)
    x2 = np.linspace(-np.pi,np.pi,256)
    y2 = np.sin(x2)
    a2.plot(x2,y2)
    a2.grid(color="g",linestyle="-.")
    
    a3 = plt.subplot(2,3,3)
    x3 = np.linspace(-np.pi,np.pi,256)
    y3 = np.cos(x3)
    a3.plot(x3,y3)
    a3.grid(color="b", linestyle=":")
    
    a4 = plt.subplot(2,3,4)
    x4 = np.arange(5)
    y4 = x4 + 5
    a4.plot(x4,y4)
    a4.grid(color="r",linestyle="--")
    
    a5 = plt.subplot(2,3,5)
    x5 = np.linspace(-np.pi,np.pi,256)
    y5 = np.sin(x5)
    a5.plot(x5,y5)
    a5.grid(color="g",linestyle="-.")
    
    a6 = plt.subplot(2,3,6)
    x6 = np.linspace(-np.pi,np.pi,256)
    y6 = np.cos(x6)
    a6.plot(x6,y6)
    a6.grid(color="b", linestyle=":")
fun()
```

```python
#保存图片
fun()
#fname:保存的图片名称
#dpi:分辨率（每英寸上的点数），默认100
plt.savefig(fname="pic.jpg",dpi=500)
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-3.5,3.5,100)
#x = [1,2,3,4]
#y = [5,6,7]
#将传入的参数x,y进行扩展，以满足3d的需求
X,Y = np.meshgrid(x,y)
Z = 3 + 0.5*np.sin(X)*np.cos(Y)-np.cos(2-X)
#绘制3d图
plt.figure(figsize=(10,9))
#加入projection="3d"说明要画3d图
a1 = plt.subplot(1,2,1,projection="3d")
#画图
a1.plot_surface(X,Y,Z)

x2 = np.arange(-4,4,0.25)
y2 = np.arange(-4,4,0.25)
X2, Y2 = np.meshgrid(x2,y2)
R = np.sqrt(X2 ** 2 + Y2 ** 2)
Z2 = np.sin(R)
a2 = plt.subplot(1,2,2,projection="3d")
s = a2.plot_surface(X2,Y2,Z2,cmap="rainbow")
plt.colorbar(s,shrink=0.5)
```
