```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
def fun1():
    price = [15000,23000, 30000, 50000]
    #第一个参数是x轴坐标，第二个参数是数据
    plt.bar(np.arange(4),price,align="center",color="steelblue",alpha=0.8)
    plt.xticks(np.arange(4),["BJ","SH","SZ","WH"])
    plt.title("Price")

fun1()    
```

```python
def fun2():
    price = [15000,23000, 30000, 50000]
    #第一个参数表示y轴的坐标，第二个参数表示x轴的值
    plt.barh(np.arange(4),price,align="center",color="steelblue")
    plt.xlim([0,70000])
    for x,y in zip(range(5),price):
        plt.text(y,x,y)
fun2()
```

```python
def fun3():
    y2016 = [13500,14500,25000,40000]
    y2017 = [16000,24000,37000,55000]
    labels = ["BJ","SZ","SH","WH"]
    #每个柱子的宽度
    bar_width = 0.45
    
    plt.bar(np.arange(4),y2016,label="2016",color="steelblue",alpha=0.8,width=bar_width)
    plt.bar(np.arange(4)+bar_width,y2017,label="2017",align="center",color="indianred",alpha=0.8,width=bar_width)
    
    plt.ylim([0,70000])
    for x,y in zip(range(4),y2016):
        plt.text(x,y,y)
        
    for x,y in zip(range(4),y2017):
        plt.text(x+bar_width,y,y)
fun3()
```

```python
def fun4():
    edu = [0.2515,0.3724,0.3336,0.0368,0.0057]
    labels = ["中专","大专","本科","硕士","其它"]
    colors = ["#9999ff","#ff9999","#7777aa","#2442aa","#dd5555"]
    
    #将横、纵坐标做标准化处理，保证饼图是正圆的
    plt.axes(aspect="equal")
    #用来突出显示大专
    explode = [0,0.1,0,0,0]
    
    plt.xlim([0,4])
    plt.ylim([0,4])
    
    plt.pie(x=edu,#绘图数据
            labels=labels,#数据的标签
           colors=colors,#饼图的颜色
           autopct="%.1f%%",#设置百分比，保留1位小数
           startangle=180,#设置图的初始角度
           frame=1,#是否显示图框
            center=(2,2),#设置饼图的原点
           explode=explode,#设置突出显示大专
           radius=1.5#设置饼的半径
           )
fun4()
```
