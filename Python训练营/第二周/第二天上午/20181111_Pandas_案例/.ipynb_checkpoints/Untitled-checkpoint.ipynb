```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
path = "orders.tsv"
df = pd.read_csv(path, sep="\t")
```

```python
#每一件商品被下单了多少次？
#df["item_name"]
#df.item_name
df.item_name.value_counts()#value_counts()获取某列中的每一项内容，在数据集中出现了多少次
```

```python
#一共有多少种商品被下单了？
```

```python
df.item_name.value_counts().shape[0]#这种办法取巧，不直接
```

```python
df.item_name.nunique()#计算某一列中不重复内容的个数
```

```python
#被下单商品的总数量是多少？
df.quantity.sum()
```

```python
#将item_price中的数据进行处理，去除掉前面的“$”，并转换成浮点数
```

```python
def str2Float(x):
    #x -> "$2.13" -> 2.13
    return float(x[1:])
#df.item_price.apply(str2Float) 将item_price中的每一个值带入apply中的函数进行计算并返回值
#df["item_price_float"] = df.item_price.apply(str2Float)
#如果不使用上一行，可以选择使用lambda表达式
df["item_price_float"] = df.item_price.apply(lambda x:float(x[1:]))
```

```python
#计算整个数据集中，所有订单的总收入是多少？
```

```python
df.item_price_float.sum()
```

```python
#当前数据集中共有多少订单？
df.order_id.nunique()
```

```python
df.order_id.value_counts().count()
```

```python
#计算数据集中每一单的平均价
#groupby分组，在分组后如果使用sum()求各，mean()平均值这类的函数
#会自动识别哪些列可以进行运算
df.groupby(["order_id"]).sum().mean()["item_price_float"]
```

```python
#获取每种英雄每个属性的平均值，并且小数点保留两位
```

```python
path = "Euro_2012_stats_TEAM.csv"
football = pd.read_csv(path)
```

```python
football
```

```python
football.shape
```

```python
football.columns
```

```python
#2012年欧洲杯总共进了多少球？
football.Goals.sum()
```

```python
#总共有多少支球队参加了2012欧洲杯？
football.shape[0]
```

```python
football.Team.count()
```

```python
#将每支球队的红牌、黄牌单独存放在一个数据集，然后进行操作
cards = football[["Team", "Red Cards", "Yellow Cards"]]
```

```python
#cards中的数据，先按照红牌排序，再按照黄牌排序
cards.sort_values(by=["Red Cards","Yellow Cards"],ascending=False)
```

```python
#计算每个球队，拿到的黄牌数量的平均值？
round(cards["Yellow Cards"].mean(),2)
```

```python
#进球数大于6的球队
football[football.Goals>6]
```

```python
#选取球队名称是以字母G开头的，  DataFrame中的字符串操作
football[football.Team.str.startswith("G")]
```

```python
#选取前7列的数据
football.ix[:,0:7]
```

```python
football.iloc[:,0:7]
```

```python
#选取除了后3列之外的全部数据
football.iloc[:,:-3]
```

```python
#获取英格兰(England)、意大利(Italy)、俄罗斯(Russia)的射正率(Shooting Accuracy)
```

```python
football.loc[football.Team.isin(["England","Italy","Russia"]),["Team","Shooting Accuracy"]]
```
